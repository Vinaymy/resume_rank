```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np

# For data structures and analysis
import pandas as pd

# Interactive vizualizations
import plotly 
import plotly.graph_objs as go
from plotly import tools
plotly.offline.init_notebook_mode(connected = True)
```

```python
df = pd.read_csv('test.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Percentage Match Must Have'] = df['Percentage Match Must Have'].str.rstrip('%').astype('float') / 100.0
df['Percentage Match Optional'] = df['Percentage Match Optional'].str.rstrip('%').astype('float') / 100.0
```

```python
df.head()
```

```python
x = df['Cosine Similarity Must Have Keywords']
data = [go.Histogram(x=x)]

layout = go.Layout(
    title='Cosine Similarity Required Keywords',
)
fig = go.Figure(data=data, layout=layout)

plotly.offline.iplot(fig)
```
